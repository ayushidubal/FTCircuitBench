```python
from qiskit import QuantumCircuit
from qiskit.visualization import circuit_drawer
from ftcircuitbench.parser.qasm_parser import transpile_qasm_to_target_basis
from ftcircuitbench.transpilers.gs_transpiler import transpile_to_gridsynth_clifford_t
from ftcircuitbench.transpilers.sk_transpiler import transpile_to_solovay_kitaev_clifford_t
```

```python
# Path to your QASM file
qasm_path = "qasm/example_circuit_for_analysis_3q.qasm"  # Change as needed

# 1. Load and draw the QASM file directly
print("Drawing original QASM circuit...")
original_circuit = QuantumCircuit.from_qasm_file(qasm_path)
circuit_drawer(original_circuit, output='mpl')
```

```python
# 2. Load and draw the circuit after processing by qasm_parser
print("Drawing processed circuit (via qasm_parser)...")
processed_circuit = transpile_qasm_to_target_basis(qasm_path, is_file=True)
processed_circuit_2 = processed_circuit.copy()
circuit_drawer(processed_circuit, output='mpl')
```

```python
# 3. After Gridsynth pipeline
gs_circuit = transpile_to_gridsynth_clifford_t(processed_circuit)
circuit_drawer(gs_circuit, output='mpl')
```

```python
# 4. After Solovay-Kitaev pipeline
sk_circuit = transpile_to_solovay_kitaev_clifford_t(processed_circuit_2)
circuit_drawer(sk_circuit, output='mpl')
```
